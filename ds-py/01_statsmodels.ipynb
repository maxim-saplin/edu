```python
import pandas as pd
data = pd.read_csv('datasets/Advertising.csv',index_col = 0)
data.head()
```

```python
model = smf.ols(formula='Q("Sales ($)") ~ Q("TV Ad Budget ($)")', data=data)
model = model.fit()
model.summary()
```

```python
print(model.params)
print(model.conf_int())
print(model.pvalues)
print(model.rsquared)
```

```python
model = smf.ols (formula = 'Q("Sales ($)") ~ Q("TV Ad Budget ($)") + Q("Radio Ad Budget ($)") + Q("Newspaper Ad Budget ($)")' ,data = data). fit()
model. summary()
```

```python
print(model.params)
print(model.conf_int())
print(model.pvalues)
print(model.rsquared)
```
