Let's assume that the data we generate represents the height of the population of age group 16-24 in Canada.
Let's fit a curve to it and perform statistical analysis on it.

Objective:
To fit a line curve on random data by providing initial guesses.

Perform the following steps
1. Generate 1100 data points between 90 and 200 using numpy arrange function.
2. Find the mean and the standard deviation of the array.
3. Create random variants of the data using norm.rvs
4. Fit a curve (line) on the data using scipy.curve_fit
5. Check how it fits, by plotting it against the

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import norm
from scipy.optimize import curve_fit
import warnings
warnings. filterwarnings( 'ignore')
%matplotlib inline
```

```python
steps = (200-90) /1100
print(steps)
```

```python
arr = np.arange(90,200, steps)
print (arr)
```

```python
plt.figure(figsize = (10,7), dpi = 100)
plt.title('Line Plot of the Array')
plt.plot(arr)
```

```python
len(arr)
```

```python
np.mean(arr)
```

```python
np.std(arr)
```

```python
data = norm.rvs(arr)
data
```

```python
axis = np.linspace(1,1100,1100)
plt.figure(figsize = (10,7), dpi = 100)
plt.title("Scatter Plot")
sns.set_style('darkgrid')
sns.scatterplot( data)
```

```python
# Fitting function
def func(x,m, c):
    return m*x + c

popt, pcov = curve_fit(func, axis, data, p0=[0,40])
popt
```

```python
pcov
```

```python
y_pred = func(axis, popt [0], popt[1])
plt.figure(figsize = (10,7) , dpi = 100 )
plt.title('Checking how good our line fits.')
plt.scatter(data, axis)
plt.plot (y_pred, axis, color = 'r')
```
