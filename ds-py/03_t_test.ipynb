- Null Hypothesis is mean of A is qeual to the mean of B
- Alternative hypothesis - not equal

```python
import random

a = [random.gauss(50,20) for x in range(30)]
b = [random.gauss(55,15) for x in range(30)]
```

```python
import seaborn

seaborn.set_style('darkgrid')
seaborn.kdeplot(a, fill=True)
seaborn.kdeplot(b, fill=True)
```

```python
import scipy.stats as stats

t_stat, p_value = stats.ttest_ind(a,b,equal_var=False) # standard deviation and mean values are not the same, we no variance is not equalt, testing for it
t_stat, p_value
```

p values is greater than 0.05, accepting the null hypothesis

```python
import numpy as np

np.mean(a), np.mean(b)
```

```python
H0 is difference between means is 0
```

```python
a = [random.gauss(50,20) for x in range(30)]
b = [random.gauss(60,25) for x in range(30)]

t_stat, p_value = stats.ttest_rel(a,b) 
t_stat, p_value
```

Rejecting, p < 0.05

```python
np.mean(a) - np.mean(b)
```
